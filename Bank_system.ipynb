```python
import time

class bank:
    
    banks_counter=0
    def __init__(self):
        self.users=[]
        self.customers=[]
        self.trades=[]
        self.workers=[]
        self.income=0
        bank.banks_counter+=1
    def create_account(self,username,password,age,id,work_status=False):
        seconds=time.time()
        date=time.ctime(seconds)
        exist=False
        for i in self.users:
            if i["id"]==id or i["username"]==username:
                exist=True
        if not exist:
            dict={
                "username":username,
                "password":password,
                "age":age,
                "id":id,
                "worker":work_status,
                "date_joined":date,
                "balance":0
            }
            self.users.append(dict)
            if work_status:
                self.workers.append(dict)
            else:
                self.customers.append(dict)
        else:
                print("user is exist")
                
    def deposit(self,username):
        dep=int(input("enter the balance you want to deposit"))
        for i in self.customers:
            if i["username"]==username:
                user=i
                break
        user["balance"]+=dep
        self.income+=dep
        
        
    def withdraw(self,username):
        w=int(input("enter the balance you want to deposit"))
        for i in self.customers:
            if i["username"]==username:
                user=i
                break
        if int(user["balance"]) >= w:
            user["balance"]-=w
            self.income-=w
            
        
    def login(self,username,password):
        
                 logined_in=False
                 for i in self.users:
                     if i["username"]==username and i["password"]==password:
                         logined_in=True
                         user=i
                         print(f"Hello {username}")
                         break
                 if logined_in:
                     while True:
                         inp=int(input("""(1) check balance
                                          (2) Deposit
                                          (3) Withdraw
                                          (4) History
                                          (5) cancel
                                          """))
                         if inp==1:
                             print(user["balance"])
                         elif inp==2:
                             seconds=time.time()
                             date=time.ctime(seconds)
                             self.deposit(user["username"])
                             dict={"user":user,
                                   "date":date,
                                   "type": "deposit"}
                             self.trades.append(dict)
                         elif inp==3:
                             seconds=time.time()
                             date=time.ctime(seconds)
                             self.withdraw(user["username"])
                             dict={"user":user,
                                   "date":date,
                                   "type": "withdraw"}
                             self.trades.append(dict)
                         elif inp==4:
                             return [trade for trad in self.trades if trade["user"]==user]
                                 
                         else:
                             return None                    
    def admin(self,username,password):
        logined_in=False
        for i in self.users:
            if i["username"]==username and i["password"]==password and i["work_status"]:
                logined_in=True
                print(f"Hello {username}")
                break
        if logined_in:       
            print(f""" the income is : {self.income}
                       the workers : {self.workers}
                       the customers : {self.customers}
                       trades : {self.trades}""")

            
        
        
```
